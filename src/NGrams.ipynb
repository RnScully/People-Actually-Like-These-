```python
import numpy as np

import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
path = 'data/english_arr.npy'
arr = np.load(path, allow_pickle = True)
```

```python
stop_words = set(stopwords.words('english'))

my_words = set([',', # punctuation
                '.',
                '"',
                "'",
                'I',# capital versions of the stop words becasue...ugh. 
                
               ])
my_stop_words = stop_words.union(my_words)
#topics[5]
```

```python
# s = 'just a simple test about trying to be better at life and getting the stopwords out of these ngrams'
# word_tokens = word_tokenize(s)
# filtered_sentence = [w for w in word_tokens if not w in stop_words]
```

```python
origin_words = arr[:,3]
```

```python
def extract_ngrams(data, num):
    n_grams = ngrams(nltk.word_tokenize(data), num)
    return [ ' '.join(grams) for grams in n_grams]
```

```python
periods = [i.replace('.','. ') for i in origin_words]
data =[]
for rows in periods:
    word_tokens = word_tokenize(rows) 
    data.append(' '.join([w.lower() for w in word_tokens if not w.lower() in my_stop_words]))
ngram_lst =[extract_ngrams(i,2) for i in data]
dashngrams = [[grams.replace(' ','-') for grams in ngrams] for ngrams in ngram_lst]

    
```

```python
np.save('data/2grams.npy',dashngrams,)
```
